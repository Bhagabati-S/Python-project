```python
## drop missing value
dataset.dropna(axis=0, inplace=True)
```

```python
##check duplicate data
dup_data=dataset.duplicated().any()
print ("Are there any duplicate values?",dup_data)
```

```python
dataset=dataset.drop_duplicates()
dataset
```

```python
dataset.columns
```

```python
dataset.reset_index(inplace = True)
```

```python
dataset
```

```python
dataset = dataset[dataset['Size'].str.contains('Varies with device') == False]
dataset
```

```python
data_M = dataset[dataset['Size'].str.contains('M')]
data_M
```

```python
data_M['Size'] = data_M['Size'].str.extract('(\d+)')
data_M
```

```python
####CONVERT INTO MB
data_M['Size'] = data_M['Size'].astype(int)
data_M['Size'] = 1000*data_M['Size']
data_M
```

```python
data_M.dtypes
```

```python
## CONVERT INTO STRINGS
```

```python
data_k = dataset[dataset['Size'].str.contains('k')]
data_k['Size'] = data_k['Size'].str.extract('(\d+)')
data_k['Size'] = data_k['Size'].astype(int)
```

```python
data_k
```

```python
new_data = pd.concat([data_M,data_k])
new_data = new_data.reset_index()
new_data
```

```python
new_data = new_data[new_data['Reviews'].str.contains('3.0M') == False]
new_data
```

```python
new_data.dtypes
```

```python
new_data['Reviews'] = new_data['Reviews'].astype(int)
```

```python
new_data.dtypes
```

```python
new_data.columns
```

```python
new_data['Installs'] = new_data['Installs'].map(lambda x:str(x).replace('+',''))
new_data['Installs'] = new_data['Installs'].map(lambda x:str(x).replace(',',''))
new_data['Installs'] = new_data['Installs'].astype(int)
```

```python
new_data
```

```python
new_data.dtypes
```

```python
###Price field is a string and has $ symbol. Remove ‘$’ sign, and convert it to numeric.
new_data['Price'] = new_data['Price'].map(lambda x:str(x).replace('$',''))
new_data['Price'] = new_data['Price'].astype(float)
```

```python
new_data
```

```python
new_data.dtypes
```

```python
###Average rating should be between 1 and 5 as only these values are allowed on the play store.
new_data = new_data[(new_data['Rating'] > 1) & (new_data['Rating'] < 5)]
new_data
```

```python
###Reviews should not be more than installs as only those who installed can review the app. 
new_data = new_data[(new_data['Installs']) > (new_data['Reviews'])]
new_data = new_data.drop(['level_0','index'], axis = 1)
new_data
```

```python
###For free apps (type = “Free”), the price should not be >0. 
new_data_1 = new_data.query('Type=="Free" and Price == 0' )
new_data_2 = new_data.query('Type=="Paid" and Price > 0' )
new_data = pd.concat([new_data_1,new_data_2])
new_data = new_data.reset_index().drop(['index'],axis = 1)
new_data
```
